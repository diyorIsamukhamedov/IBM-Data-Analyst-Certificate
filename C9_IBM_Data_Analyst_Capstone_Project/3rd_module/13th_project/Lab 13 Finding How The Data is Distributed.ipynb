### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
## Write your code here
summary = pd.DataFrame({
    "Column name": df.columns,
    "DataType": df.dtypes,
    "Number of columns": df.shape[0],
    "Number of rows": df.shape[1],
    "Not-Null Count": df.notnull().sum(),
    "Missing Count": df.isnull().sum(),
    "Missing Percentage": df.isnull().mean() * 100
})

display(summary)
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
# Set display.max_rows to None to show all rows
pd.set_option('display.max_rows', None)
```

```python
## Write your code here
data_type = df.dtypes
missing_counts = df.isnull().sum()
missing_percent = (missing_counts / len(df)) * 100

missing_summary_v2 = pd.DataFrame({
    'Data type of the columns': data_type,
    'MissingCount': missing_counts,
    'MissingPercent': missing_percent
}).sort_values(by='MissingPercent', ascending=False)

display(missing_summary_v2)
```

```python
# We are going to delete columns, where null values < 70 inclusive.
treshold = 0.7
df = df.loc[:, df.isnull().mean() <= treshold]

# Fill in null values
for col in df.columns:
    # Fill in numeric columns that contain null values with median
    if df[col].dtype in ['float64', 'int64']:
        df.loc[:, col] = df[col].fillna(df[col].median())
    # Fill in categorical columns that contain null values with mode
    else:
        df.loc[:, col] = df[col].fillna(df[col].mode()[0])

missing_count = df.isnull().sum().sum()

if missing_count == 0:
    print('No Null values. All clear!')
else:
    print(f'There is/are some/any Null/Nulls val: {missing_count}')
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



If a column has too many missing values (e.g., >50%), it might be better to remove it or carefully consider if it can be dropped.

For numerical columns with few missing values, impute with the mean or median.

For categorical columns with few missing values, impute with the most frequent value (mode).

For rows with many missing values, you might consider dropping those rows if appropriate.

```python
## Write your code here
# Store key columns in the variable
key_columns = ['Employment','JobSat','YearsCodePro']

# We are going to delete columns, where null values < 50 inclusive.
treshold = 0.5 # 50%
row_na_fraction = df[key_columns].isnull().mean(axis=1)
df = df.loc[row_na_fraction <= treshold]

# Fill in null values
for col in key_columns:
    # Fill in numeric columns that contain null values with median
    if df[col].dtype in ['float64', 'int64']:
        df.loc[:, col] = df[col].fillna(df[col].median())
    # Fill in categorical columns that contain null values with mode
    else:
        df.loc[:, col] = df[col].fillna(df[col].mode()[0])

missing_count = df[key_columns].isnull().sum().sum()

if missing_count == 0:
    print('No Null values in key columns. All clear!')
else:
    print(f'Some Nulls remain in key columns: {missing_count}')
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here
# Mapping numerical values into categorical
job_sat_map = {
    10: "Very satisfied",
    9: "Very satisfied",
    8: "Satisfied",
    7: "Satisfied",
    6: "Neutral",
    5: "Neutral",
    4: "Dissatisfied",
    3: "Dissatisfied",
    2: "Very dissatisfied",
    1: "Very dissatisfied",
    0: "Very dissatisfied"
}

# Create new categorical column
df['JobSatCategory'] = df['JobSat'].map(job_sat_map)

# Create the pie 2 chart
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# The first pie chart (with row data)
df['JobSat'].value_counts().plot.pie(
    autopct='%1.1f%%',
    ax=axes[0],
    startangle=90
)
axes[0].set_ylabel('')
axes[0].set_title('Job Satisfaction (Numeric)')

# The second pie chart (for stakeholders)
df['JobSatCategory'].value_counts().plot.pie(
    autopct='%1.1f%%',
    ax=axes[1],
    startangle=90
)
axes[1].set_ylabel('')
axes[1].set_title('Job Satisfaction (Categorical)')

plt.tight_layout()
plt.show()
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here

# Separate rows with ; and count the frequency
have_worked_with_counts = df['LanguageHaveWorkedWith'].str.split(';').explode().value_counts()
want_to_work_with_counts = df['LanguageWantToWorkWith'].str.split(';').explode().value_counts()

# Union into 1 DataFrame
language_df = pd.DataFrame({
    'Have Worked With': have_worked_with_counts,
    'Want To Work With': want_to_work_with_counts
}).fillna(0).astype(int)

language_df.plot(kind='bar', figsize=(12, 6), width=0.8)
plt.title('Programming Languages: Have Worked With vs Want To Work With')
plt.ylabel('Number of Respondents')
plt.xlabel('Programming Language')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here
# remote_region = (
#     df.groupby(["Country", "RemoteWork"])
#     .count()
#     .pivot_table(
#         index = 'Country',
#         columns = 'RemoteWork',
#         values = 'SomeColumn',
#         fill_value=0
#     )
# )

# Output top 20 regions. The reason is because, we cannot see all regions on the chart
top_regions = df['Country'].value_counts().head(20).index

# Filter those regions
df_top = df[df['Country'].isin(top_regions)]

remote_region = pd.crosstab(df_top['Country'], df_top['RemoteWork'])

remote_region.plot(kind='bar', figsize=(12, 6), width=0.8)
plt.title('Remote Work Distribution by Region')
plt.ylabel('Number of Respondents')
plt.xlabel('Region')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
## Write your code here
df['YearsCodePro'] = (
    df['YearsCodePro']
    .replace('Less than 1 year', 0.5)
    .replace('More than 50 years', 51)
    .astype(float)
)
# Pearson correlation is my choice
pearson_corr = df['JobSat'].corr(df['YearsCodePro'], method='pearson')

print(f'Pearson correlation: {pearson_corr}')
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.
